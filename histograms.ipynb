```python
import matplotlib.pyplot as plt
import geopandas as gpd
import numpy as np
```

```python
gpkg_path = "outputs/merged_structures.gpkg"
layer_name = "merged_structures"

gdf = gpd.read_file(gpkg_path, layer=layer_name)
```

```python
avg_burned_proportion_10m = gdf["burned_proportion_10m"].mean(skipna=True)
print(f"Average burned_proportion_10m: {avg_burned_proportion_10m:.2f}")
```

```python
avg_burned_proportion_20m = gdf["burned_proportion_20m"].mean(skipna=True)
print(f"Average burned_proportion_20m: {avg_burned_proportion_20m:.2f}")
```

```python
# Extract column and drop NaNs
values = gdf["burned_proportion_10m"].dropna()

# Define histogram range (e.g., from min to max rounded to nearest whole number)
min_val = np.floor(values.min())
max_val = np.ceil(values.max())

# Create histogram
plt.figure(figsize=(8, 5))
plt.hist(values, bins=50, edgecolor='black')

plt.xlabel("Mean Burned Proportion")
plt.ylabel("Number of Polygons")
plt.title(" Distribution of Burned Proportion 10m")

plt.xlim(min_val, max_val)
plt.xticks(np.arange(min_val, max_val + 1, 1))
plt.grid(True, linestyle='--', alpha=0.6)
plt.tight_layout()
plt.show()
```
